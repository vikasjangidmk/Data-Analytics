```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
import sys
%matplotlib inline
```

```python
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=1000, n_features=2, n_redundant=0, n_clusters_per_class=1,
                    weights=[0.90],random_state=12)
```

```python
import pandas as pd

df1 = pd.DataFrame(X,columns=['f1','f2'])
df2 = pd.DataFrame(y,columns=['target'])
df = pd.concat([df1,df2],axis=1)
df.head()
```

```python
df['target'].value_counts()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.scatter(df['f1'],df['f2'],c=df['target'],cmap='viridis')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
from imblearn.over_sampling import SMOTE
```

## Transform the Dataset

```python
oversample = SMOTE()
X,y = oversample.fit_resample(df[['f1','f2']],df['target'])
```

```python
X.shape
```

```python
len(y[y==0]),len(y[y==1])
```

```python
import pandas as pd

df1 = pd.DataFrame(X,columns=['f1','f2'])
df2 = pd.DataFrame(y,columns=['target'])
oversample_df = pd.concat([df1,df2],axis=1)
oversample_df.head()
```

```python
oversample_df['target'].value_counts()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(oversample_df['f1'],oversample_df['f2'],c=oversample_df['target'],cmap='viridis')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
